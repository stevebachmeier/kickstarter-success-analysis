```python
# ========================================
# READ/EXTRACT RAW DATA
# ========================================
with open("data/Kickstarter_2018-10-18T03_20_48_880Z/Kickstarter_2018-10-18T03_20_48_880Z.json",
          encoding="utf8") as json_file:
    json_obj = [json.loads(line) for line in json_file]
```
